# relational plot 

```python
#!pip install plotly.express
import seaborn as sns 
import matplotlib as plt 
import plotly.express as px

tips = sns.load_dataset('tips')
```

```python
sns.scatterplot(data = tips,x = 'total_bill',y = 'tip',hue='sex',style = 'time')
# scatter plot - Axis level function 
```

```python
sns.relplot(data = tips ,x = 'total_bill',y = 'tip',kind = 'scatter',hue = 'sex',style = 'time')
#relplot is a figure level function 
#best is to use relplot because using figure level function we can create many axis level functions 
```

# LINE PLOT

```python
gap = px.data.gapminder()
temp_df = gap[gap['country'] =='India']
temp_df
```

```python
sns.lineplot(data = temp_df,x = 'year',y = 'lifeExp')
```

```python
sns.relplot(data = temp_df,x = 'year',y = 'lifeExp',kind = 'line')
```

```python
temp = gap[gap['country'].isin(['India','Brazil','Germany'])]
```

```python
sns.relplot(data = temp,x ='year',y = 'lifeExp',kind = 'line',hue = 'country',style = 'continent')
```

```python
sns.lineplot(data = temp,x ='year',y = 'lifeExp',hue = 'country')
```

```python
# facet plot 
sns.relplot(data = tips ,x = 'total_bill',y = 'tip',kind = 'scatter',row = 'sex')
```

```python
sns.relplot(data = tips ,x = 'total_bill',y = 'tip',kind = 'line',row = 'sex',col = 'smoker',)
```

```python
# facet plot can be used only with figure plot and not axis level plot
```

```python
sns.relplot(data = gap,x='lifeExp',y='gdpPercap',kind ='scatter',col='year',col_wrap =4)
```

# distribution plots 
1)histplot 
2)kdeplot 
3)rugplot

```python
#figure level - displot 
#axis level - histplot-->kdeplot-->rugplot
```

```python
sns.histplot(data = tips,x ='total_bill')
```

```python
sns.histplot(data = tips,x ='total_bill',bins = 20)
```

```python
tips
```

```python
sns.displot(data = tips,x ='day',kind ='hist')
```

```python
sns.displot(data = tips,x ='tip',kind ='hist',hue='sex',element='step')
```

```python
df = sns.load_dataset('titanic')
df
```

```python
sns.displot(data = df,x='age',kind='hist',hue='sex')
```

```python
#faceting happens only for figure level functions.
sns.displot(data = df,x='age',kind='hist',col='sex')
```

# kde plot (kernel density estimation)

```python
#guassian kernel producing a continuous density estimate.
```

```python
sns.kdeplot(data = tips,x='total_bill')
```

```python
sns.displot(data = df,x='age',kind='kde',hue='sex',fill=True)
```

# Rug plot

```python
sns.kdeplot(data = df,x='age')
sns.rugplot(data = df,x='age')
```
